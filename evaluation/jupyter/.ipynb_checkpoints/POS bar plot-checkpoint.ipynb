```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import json
import nltk
from collections import Counter

# nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

with open('../json_files/sequence.json') as json_file:
    data = json.load(json_file)
    
```

```python
# inserted List

insertedList = []
for value in data.values():
    for item in value:
        if(len(item['insertedList']) > 0):
            for i in item['insertedList']:
                insertedList.append(i)
                
inserted_pos_tag = nltk.pos_tag(insertedList)
inserted_pos = [i[1] for i in inserted_pos_tag]
inserted_pos

fig, ax = plt.subplots(figsize=(16, 12))

x = Counter(inserted_pos).keys()
y = Counter(inserted_pos).values()
plt.bar(x, y, color='blue', lw=5, alpha=0.8)

rb_patch = mpatches.Patch( label='RB = Adverb')
in_patch = mpatches.Patch( label='IN = Preposition or subordinating conjunction')
jj_patch = mpatches.Patch( label='JJ = Adjective')
cc_patch = mpatches.Patch( label='CC = Coordinating conjunction')
nns_patch = mpatches.Patch( label='NNS = Noun, plural')
vb_patch = mpatches.Patch( label='VB = Verb, base form')
prp_patch = mpatches.Patch( label='PRP = Personal pronoun')
nn_patch = mpatches.Patch(label='NN = Noun, singular or mass')

plt.legend(handles=[rb_patch,in_patch,jj_patch,cc_patch,nns_patch,vb_patch,prp_patch,nn_patch], frameon=True)
plt.style.use('fivethirtyeight')

ax.set_title('Inserted words', fontsize=24)
ax.set_xlabel('Part of Speech', fontsize=20)
ax.set_ylabel('# Occurance', fontsize=20)

plt.savefig('../images/pos_inserted.png')
plt.show()
```

```python
# deleted List

deletedList = []
for value in data.values():
    for item in value:
        if(len(item['deletedList']) > 0):
            for i in item['deletedList']:
                deletedList.append(i)
                
deleted_pos_tag = nltk.pos_tag(deletedList)
deleted_pos = [i[1] for i in deleted_pos_tag]
deleted_pos

fig, ax = plt.subplots(figsize=(16, 12))
x = Counter(deleted_pos).keys()
y = Counter(deleted_pos).values()
plt.bar(x, y, color='blue', lw=5, alpha=0.8)

rbr_patch = mpatches.Patch( label='RBR = Adverb, comparative')
dt_patch = mpatches.Patch( label='DT = Determiner')
nn_patch = mpatches.Patch(label='NN = Noun, singular or mass')
in_patch = mpatches.Patch( label='IN = Preposition or subordinating conjunction')
jjr_patch = mpatches.Patch( label='JJR = Adjective, comparative')
rb_patch = mpatches.Patch( label='RB = Adverb')
cc_patch = mpatches.Patch( label='CC = Coordinating conjunction')
to_patch = mpatches.Patch( label='TO = to')
nnp_patch = mpatches.Patch( label='NNP = Proper noun, singular')
nns_patch = mpatches.Patch( label='NNS = Noun, plural')
ex_patch = mpatches.Patch( label='EX = Existential there')
vbp_patch = mpatches.Patch( label='VBP = Verb, non-3rd person singular present')
jj_patch = mpatches.Patch( label='JJ = Adjective')


plt.legend(handles=[rbr_patch,dt_patch,nn_patch,in_patch,jjr_patch,rb_patch,cc_patch,to_patch,nnp_patch,nns_patch,ex_patch,vbp_patch,jj_patch], frameon=True)
plt.style.use('fivethirtyeight')

ax.set_title('Removed words', fontsize=24)
ax.set_xlabel('Part of Speech', fontsize=20)
ax.set_ylabel('# Occurance', fontsize=20)
ax.grid(True)
plt.savefig('../images/pos_removed.png')
plt.show()
```

```python
# substituted List

substitutedList = []
for value in data.values():
    for item in value:
        if(len(item['substitutedList']) > 0):
            for inner in item['substitutedList']:
                for i in inner:
                    substitutedList.append(i)
                
substituted_pos_tag = nltk.pos_tag(substitutedList)
substituted_pos = [i[1] for i in substituted_pos_tag]
substituted_pos

fig, ax = plt.subplots(figsize=(16, 12))

x_substituted_pos = Counter(substituted_pos).keys()
y_substituted_pos = Counter(substituted_pos).values()

plt.bar(x_substituted_pos, y_substituted_pos, color='blue', lw=5, alpha=0.8)

xlocs, xlabs = plt.xticks()
# for i, v in enumerate(y):
#     plt.text(xlocs[i] - 0.1, v + 0.7, str(v))

rbr_patch = mpatches.Patch( label='RBR = Adverb, comparative')
rb_patch = mpatches.Patch( label='RB = Adverb')
in_patch = mpatches.Patch( label='IN = Preposition or subordinating conjunction')
nn_patch = mpatches.Patch(label='NN = Noun, singular or mass')
nns_patch = mpatches.Patch( label='NNS = Noun, plural')
dt_patch = mpatches.Patch( label='DT = Determiner')
jj_patch = mpatches.Patch( label='JJ = Adjective')
jjr_patch = mpatches.Patch( label='JJR = Adjective, comparative')
to_patch = mpatches.Patch( label='TO = to')
vbg_patch = mpatches.Patch( label='VBG = Verb, gerund or present participle')
vb_patch = mpatches.Patch( label='VB = Verb, base form')
nnp_patch = mpatches.Patch( label='NNP = Proper noun, singular')
md_patch = mpatches.Patch( label='MD = Modal')
cc_patch = mpatches.Patch( label='CC = Coordinating conjunction')
vbp_patch = mpatches.Patch( label='VBP = Verb, non-3rd person singular present')
wdt_patch = mpatches.Patch( label='WDT = Wh-determiner')


plt.legend(handles=[rbr_patch,rb_patch,in_patch,nn_patch,nns_patch,dt_patch,jj_patch,jjr_patch,to_patch,vbg_patch,vb_patch,nnp_patch,md_patch,cc_patch,vbp_patch,wdt_patch], frameon=True)
plt.style.use('fivethirtyeight')

ax.set_title('Substituted words', fontsize=24)
ax.set_xlabel('Part of Speech', fontsize=20)
ax.set_ylabel('# Occurance', fontsize=20)
ax.grid(True)
plt.savefig('../images/pos_substituted.png')
plt.show()
```

```python
def make_pos(str):
    tokens = nltk.word_tokenize(str)
    pos_tag = nltk.pos_tag(tokens)
    pos = [i[1] for i in pos_tag]
    return pos

def make_bigrams(pos):
    return list(nltk.bigrams(pos))

for value in data.values():
    for item in value:
        item['pos_original'] = make_pos(item['originalSentence'])
        item['pos_modified'] = make_pos(item['modifiedSentece'])    

def get_diff(lst1, lst2):
    return tuple(set(lst1) - set(lst2))
```

```python
pos_dif = []
sequenced_bigrams = []
for value in data.values():
    for item in value:
        item['bigrams'] = get_diff(make_bigrams(item['pos_original']), make_bigrams(item['pos_modified']))
        sequenced_bigrams.append(get_diff(make_bigrams(item['pos_original']), make_bigrams(item['pos_modified'])))
        pos_dif.append(get_diff(item['pos_original'],item['pos_modified']))
```
