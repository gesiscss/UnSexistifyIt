```python
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from collections import Counter
import nltk
nltk.download('punkt')

with open('../json_files/prepared_sentences.json') as json_file:
    data = json.load(json_file)
    
data
```

```python
rating_list = []
for value in data.values():
    for item in value:
        rating_list.append(item['rating'])
        
rating_list
```

```python
unique_rating = Counter(rating_list).keys()
unique_rating_count = Counter(rating_list).values()

unique_rating_dict = dict(zip(unique_rating, unique_rating_count))
unique_rating_dict = dict(unique_rating_dict.items(), key=lambda kv: kv[1])
unique_rating_dict
```

```python
fig, ax = plt.subplots(figsize=(16, 12))

x = unique_rating
y = unique_rating_count
plt.bar(x, y, color='blue', lw=5, alpha=0.8)

one_patch = mpatches.Patch( label='1 = NON-SEXIST')
five_patch = mpatches.Patch( label='5 = SEXIST')

plt.legend(handles=[one_patch,five_patch], frameon=True)

plt.style.use('fivethirtyeight')

ax.set_title('Modified Sentences Rating', fontsize=24)
ax.set_xlabel('Rates', fontsize=20)
ax.set_ylabel('Number of rates', fontsize=20)

plt.savefig('../images/rating_bar.png')
plt.show()
```
