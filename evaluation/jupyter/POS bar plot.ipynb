```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import json
import nltk
from collections import Counter

nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

with open('../json_files/sequence.json') as json_file:
    data = json.load(json_file)
    
# data    
```

```python
def make_pos(str):
    tokens = nltk.word_tokenize(str)
    pos_tag = nltk.pos_tag(tokens)
    pos = [i[1] for i in pos_tag]
    return pos

def make_bigrams(pos):
    return list(nltk.bigrams(pos))

for value in data.values():
    for item in value:
        item['pos_original'] = make_pos(item['originalSentence'])
        item['pos_modified'] = make_pos(item['modifiedSentece'])    

def get_diff(lst1, lst2):
    return tuple(set(lst1) - set(lst2))
```

```python
pos_dif = []
sequenced_bigrams = []
for value in data.values():
    for item in value:
        item['bigrams'] = get_diff(make_bigrams(item['pos_original']), make_bigrams(item['pos_modified']))
        sequenced_bigrams.append(get_diff(make_bigrams(item['pos_original']), make_bigrams(item['pos_modified'])))
        pos_dif.append(get_diff(item['pos_original'],item['pos_modified']))
```
