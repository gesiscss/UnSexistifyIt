```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import json

with open('../json_files/cdf_data.json') as json_file:
    data = json.load(json_file)
```

```python
data
```

```python
levenshtein = []
for value in data.values():
    for item in value:
        levenshtein.append(item['levenshtein'])

fig, ax = plt.subplots(figsize=(16, 12))

x = np.sort(levenshtein)
y = np.arange(len(x))/float(len(x))

# plot(x, y, linestyle='dashed')

# Plot both
plt.style.use('fivethirtyeight')
plt.plot(x, y , color='black', lw=5, alpha=0.8)
ax.grid(True)
ax.set_title('Levenshtein Distance', fontsize=24)
ax.set_xlabel('Value', fontsize=20)
ax.set_ylabel('Sentences (%)', fontsize=20)
plt.savefig('../images/levenshtein.png')
plt.show()
```

```python
lenght_diff = []
for value in data.values():
    for item in value:
        lenght_diff.append(item['lenght_diff'])
        
fig, ax = plt.subplots(figsize=(16, 12))

x = np.sort(lenght_diff)
y = np.arange(len(x))/float(len(x))

# plt.style.use('fivethirtyeight')
plt.plot(x,y , color='black', lw=5, alpha=0.8)

ax.grid(True)
ax.set_title('Difference in Number of Words', fontsize=24)
ax.set_xlabel('Value', fontsize=20)
ax.set_ylabel('Sentences (%)', fontsize=20)
plt.savefig('../images/length_diff.png')
plt.show()
```
